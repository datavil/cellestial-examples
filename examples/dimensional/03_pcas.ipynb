```python
import cellestial as cl
import scanpy as sc
from lets_plot import *

LetsPlot.setup_html()

data = sc.read("data/pbmc3k_mini.h5ad")
```

## `pcas` is a multi-grid plot function and accepts multiple keys list or any iterable.

```python
umap_grid = (
    cl.pcas(
        data,
        keys=["leiden", "HBD", "NEAT1", "IGKC"],
        ncol=2,
        size=0.8,
        color_high="#D2042D",  # works with hex codes, rgb codes, and names (red, blue, etc.)
    )
    + ggsize(900, 700)
)
umap_grid
```
