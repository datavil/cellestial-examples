```python
import cellestial as cl
import scanpy as sc
from lets_plot import *

LetsPlot.setup_html()

data = sc.read("data/pbmc3k_pped.h5ad")
```

### A simple `tSNE` plot

```python
plot = cl.tsne(data, key="leiden", axis_type="arrow", size=0.6, interactive=True)
plot
```

## Of course, it is possible to customize the plot

```python
plot + scale_color_hue()
```

## You can also use a gene a key.

```python
cl.tsne(data, "NEAT1", color_high="#D2042D")
```
